```python
import shutil
from google.colab import drive
import os

# Mount Google Drive
drive.mount('/content/drive')

# Define the source and destination paths
notebook_path = '/content/train_model_online.ipynb'  # Replace with your notebook's name
repo_path = '/content/drive/My Drive/SODeep/train_model_online.ipynb'

# Copy the notebook
if os.path.exists(notebook_path):
    shutil.copy(notebook_path, repo_path)
    print(f"Notebook copied to: {repo_path}")

    # Optionally delete the original file
    os.remove(notebook_path)
    print(f"Original file deleted from: {notebook_path}")
else:
    print(f"Notebook not found at: {notebook_path}")

```
